```python
# import libraries
import seaborn as sns
import pandas as pd
import numpy as np

# load data_set
kashti=sns.load_dataset("titanic")
kashti.head(4)    # To see the data structure of the data

p2=sns.boxplot(x="survived",y="age",data=kashti,showmeans=True,hue="who")  # Shows mean in the box plot
```

```python
# import libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# load data_set
kashti=sns.load_dataset("titanic")
kashti.head(4)    # To see the data structure of the data

sns.boxplot(x="survived",y="age",showmeans=True,
            meanprops={"marker":"*","markersize":"12","markeredgecolor":"red"},hue="who",data=kashti,)

# show labels
plt.xlabel=('How many survived',size=10,weight='bold')
plt.ylabel('Age(years)')
plt.title=("Box plot of titanic")
plt.show()
```
