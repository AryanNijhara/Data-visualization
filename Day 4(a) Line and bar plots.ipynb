```python
#draw line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.show()
```

```python
#Adding titles
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("phoolon ka plot")
plt.show()
```

```python
#Adding limits
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("phoolon ka plot")
plt.xlim(2)
plt.ylim(1)

plt.show()
```

```python
# change location of male and female
#load_dataset
kashti=sns.load_dataset("titanic")
kashti

#draw bar plot
sns.barplot(x="who",y="alone",hue="who",data=kashti, order=["woman","man"])
plt.show()
```

```python
#Colour changing
# change location of male and female
#load_dataset
kashti=sns.load_dataset("titanic")
kashti

#draw bar plot
sns.barplot(x="who",y="alone",hue="who",data=kashti, order=["woman","man"],color="brown")
plt.show()
```

```python
# Removing error bars and palette(colours)
#Colour changing
# change location of male and female
#load_dataset
kashti=sns.load_dataset("titanic")
kashti

#draw bar plot
sns.barplot(x="who",y="alone",hue="who",data=kashti, order=["woman","man"],color="brown",ci=None,palette="pastel")
plt.show()
```

```python
#estimator of median
from numpy import median
#load_dataset
kashti=sns.load_dataset("titanic")
kashti

#draw bar plot
sns.barplot(x="class",y="fare",hue="who",data=kashti,estimator=median,saturation=0.1)
plt.show()
```

```python
#estimator of median
from numpy import median
import seaborn as sns
import matplotlib.pyplot as plt
#load_dataset
kashti=sns.load_dataset("titanic")
kashti

#draw bar plot
sns.barplot(x="fare",y="class",hue="who",data=kashti,estimator=median,saturation=1)
plt.show()
```

## colour of error and bars

```python
#estimator of median
from numpy import median
import seaborn as sns
import matplotlib.pyplot as plt
#load_dataset
kashti=sns.load_dataset("titanic")
kashti

#draw bar plot
sns.barplot(x="class",y="fare",hue="who",data=kashti,estimator=median,saturation=0.1,
           linewidth=3,facecolor=(1,0.7,1,0),errcolor=".5",edgecolor="0")
plt.show()
```
